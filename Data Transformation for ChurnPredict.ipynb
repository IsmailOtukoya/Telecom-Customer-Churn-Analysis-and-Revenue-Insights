```python
import pandas as pd

df  = pd.read_csv(r"C:/Users/ISMAIL/OneDrive/Documents/data science projects/cleanedandanalyzed_dataset.csv")
```

```python
df.head()
```

```python
#Data type conversion
#Convert "Gender," "Married," "City," "Offer," and other categorical columns from object to the appropriate data type (e.g., category or boolean).
#Convert "Avg Monthly GB Download" from object to a numeric data type (e.g., float) for easier analysis.

df['Gender'] = df['Gender'].astype('category')
df['Married'] = df['Married'].astype('category')
df['City'] = df['City'].astype('category')
df['Offer'] = df['Offer'].astype('category')
df['Avg Monthly GB Download'] = pd.to_numeric(df['Avg Monthly GB Download'], errors='coerce') 
df['Gender'] = df['Gender'].astype('category')
df['Married'] = df['Married'].astype('category')
df['City'] = df['City'].astype('category')
df['Offer'] = df['Offer'].astype('category')
df['Avg Monthly GB Download'] = pd.to_numeric(df['Avg Monthly GB Download'], errors='coerce')
```

```python
df.dtypes
```

```python
# Creating a new column for age group based on the 'Age' column
df['Age Group'] = pd.cut(df['Age'], bins=[0, 17, 24, 34, 44, 54, 64, 200], labels=['0-17', '18-24', '25-34', '35-44', '45-54', '55-64', '65+'])

# Creating a binary variable indicating whether a customer has dependents
df['Has Dependents'] = df['Number of Dependents'].apply(lambda x: 1 if x > 0 else 0)

# Calculating the total charges per month
df['Total Charges Per Month'] = df['Monthly Charge'] + df['Total Charges'] - df['Total Refunds'] + df['Total Extra Data Charges'] + df['Total Long Distance Charges']

# Display the updated DataFrame
print(df)
```

```python
# Select the numerical columns for standardization
numeric_columns = ['Monthly Charge', 'Total Charges', 'Total Refunds', 'Total Extra Data Charges', 'Total Long Distance Charges']

# Perform z-score normalization on the selected columns
df[numeric_columns] = (df[numeric_columns] - df[numeric_columns].mean()) / df[numeric_columns].std()

```
